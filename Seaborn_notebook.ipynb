# Seaborn

### Why using seaborn while having matplotlib?

1. seaborn makes your charts prettier.
2. Most of the customisations that work for Matplotlib work for Seaborn.
3. Seaborn uses fewer syntax and has stunning default themes. (while Matplotlib is more easily customizable).

#### Seaborn is a graphic library built on top of Matplotlib.

despite all of this, most of your work in data visualization will combine between these two libraries.

 - https://youtu.be/9jYqZS142mg

 - https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/interquartile-range-iqr/a/interquartile-range-review

## 1.Quartiles:
- four Values that divide your data into quarters (4 equal parts), denoted as Q1, Q2, Q3, Q4. Such that:
 - Q1: __25%__ of the data is less than __Q1.__
 - Q2: __50%__ of the data is less than __Q2.__ __It is also known the medien.__
 - Q3: __75%__ of the data is less than __Q3.__
 - Q4: __All__ the data is less than __Q4.__ __Hence, It is also known as the Maximum.__

## 2.Interquartiles (IQR):
- The difference between the 3rd and 1st quartiles __( IQR = Q3 - Q1 )__.
- Why do we need IQR?
 - Used to build box plots.
 - It is one way to find outliers.

# Outliers:
 - https://youtu.be/9jYqZS142mg
 - __Def1:__ An outlier is an extremely high or low data value compared with the other data values in the dataset.
 - __Def2:__ An outlier is an observation that lies an abnormal distance from other values in a random sample from a population. 
 - In a sense, this definition leaves it up to the analyst to decide what will be considered abnormal.
 
 - Outliers affect the statistics values of your dataset (like mean and standard deviation)
 - Outliers can be found using the following formula:
  - Outlier= 
    - Any value smaller than Q1 - 1.5 * IQR.
    - Any value bigger than Q3 + 1.5 * IQR.

# Boxplot:
 - A box plot shows the five number summary of a dataset:
   - __Minimum value.__
   - __Lower quartile (Q1).__
   - __Median (Q2).__
   - __Upper quartile (Q3).__
   - __Maximum value.__


### You can draw a box plot for a feature in your dataset using the seaborn function 'boxplot'.
#### sns.boxplot(x= )
https://seaborn.pydata.org/generated/seaborn.boxplot.html
